## 2. Most common 15 Name or Surname of killed people 

```python
kill.head()
```

```python
kill.shape
```

```python
kill.name.value_counts()
```

```python
# Problem: TK TK is most probably not a name or surname. We will clean them.
pairs = kill.name[kill.name != 'TK TK'].str.split()
pairs.head()
```

**INSTRUCTION : Plot Most common 15 Name or Surname of killed people** 

```python
count_2=0
count_3=0
count_4=0
for i in pairs:
    if len(i) == 2:
        count_2 += 1
    elif len(i) == 3:
        count_3 += 1
    elif len(i) == 4:
        count_4 += 1
print('count_2:', count_2)
print('count_3:', count_3)
print('count_4:', count_4)
```

```python
len(pairs) == count_2 + count_3 + count_4
```

```python
a=[]
b=[]
c=[]
d=[]

for i in pairs:
    if len(i) == 2:
        a.append(i[0])
        d.append(i[1])
    elif len(i) == 3:
        a.append(i[0])
        b.append(i[1])
        d.append(i[2])
    elif len(i) == 4:
        a.append(i[0])
        b.append(i[1])
        c.append(i[2])
        d.append(i[3])
            
```

```python
pairs
```

```python
name_total = a + b + c
```

```python
name_count=Counter(name_total)
surname_count=Counter(d)
```

```python
name_count.most_common(13)
```

```python
surname_count.most_common(2)
```

```python
most_com_total = name_count.most_common(13) + surname_count.most_common(2)
```

```python
most_com_total.sort(reverse=True)
```

```python
most_com_total[0][1]
```

```python
most_com_total_sorted = sorted(most_com_total, key=lambda x: x[1], reverse=True)
```

```python
most_com_total_sorted
```

```python
x,y = zip(*most_com_total_sorted)
```

```python
x
```

```python
y
```

```python
type(x)
```

```python
x,y = list(x), list(y)
```

```python
# Plotting with Seaborn
plt.figure(figsize=(15,8))
g=sns.barplot(x,y)
plt.title("Most Common 15 Name or Surname")
plt.xlabel("Name or Surname of Killed People")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
legend = True
for p in g.patches:
    g.annotate((p.get_height()), (p.get_x()+0.2, p.get_height()+0.5));
plt.show()
```

**INSTRUCTION-2: Demonstrate Race Ratio of Killed People by a Pie Chart**

```python
kill['race'].value_counts()
```

```python
kill.groupby('race')['race'].count().sort_values(ascending=False).index
```

```python
kill.groupby('race')['race'].count().sort_values(ascending=False).values
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = kill.groupby('race')['race'].count().sort_values(ascending=False).index
sizes = kill.groupby('race')['race'].count().sort_values(ascending=False).values
explode = (0, 0, 0, 0,0,0.2)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots(figsize=(14,10))
ax1.pie(sizes, explode=explode, labels=labels, shadow=False,autopct='%1.1f%%', startangle=0)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_title('Killed People According to Races', fontdict={'color':'blue', 'fontsize':30})
plt.show()
```
